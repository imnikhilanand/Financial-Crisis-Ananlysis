### Correlation between different features

```python
# correlation between different features in the data
data.corr()
plt.figure(figsize=(12,6))
sns.heatmap(data.corr(), annot=True, cmap='Blues')
plt.show()
```

Observations:

- There is slight correlation between year and indepedence. Since, most the countries got independence as the time groups.
- There is slight correlation between exch_usd and and sovereign_external_debt_default.
- There is slight correlation between domestic_debt_in_default and sovereign_external_debt_default.
- There is slight correlation between currency_crises and inflation_crises.
